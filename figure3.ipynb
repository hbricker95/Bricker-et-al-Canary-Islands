### defining skill-score equation

```python
import numpy as np

def find_nearest_index(array, value):
    return np.abs(array - value).argmin()

def compute_skill_score_error(oi, ei, mi):
    oi = np.array(oi)
    ei = np.array(ei)
    mi = np.array(mi)
    numerator = np.sum((mi - oi) ** 2) - np.sum(ei ** 2)
    denominator = np.sum((0 - oi) ** 2) - np.sum(ei ** 2)
    skill_score = 1 - np.sqrt(numerator / denominator)
    return skill_score

def compute_skill_score_no_error(oi, ei, mi):
    oi = np.array(oi)
    ei = np.array(ei)
    mi = np.array(mi)
    numerator = np.sum((mi - oi) ** 2)
    denominator = np.sum((0 - oi) ** 2) 
    skill_score = 1 - np.sqrt(numerator / denominator)
    return skill_score

def extract_mi_from_data(latitudes_nc, longitudes_nc, mi_data, latitudes, longitudes):
    mi_values = []

    for lat, lon in zip(latitudes, longitudes):
        lat_idx = find_nearest_index(latitudes_nc, lat)
        lon_idx = find_nearest_index(longitudes_nc, lon)
        # Extract mi for the nearest latitude and longitude from the pre-loaded data array
        mi_values.append(mi_data[lat_idx, lon_idx])

    return np.array(mi_values)

# Scenario 1: Using pre-loaded nc_data variables directly (no file path)
def skill_score_scenario_error(nc_data_list, P_anom, P_clumped_error, latties, longies):
    skill_scores = []  # List to store the skill scores
    for i, (latitudes_nc, longitudes_nc, mi_data) in enumerate(nc_data_list):
        mi = extract_mi_from_data(latitudes_nc, longitudes_nc, mi_data, latties, longies)
        skill_score = compute_skill_score_error(P_anom, P_clumped_error, mi)
        skill_scores.append(skill_score)  # Store skill score in the list
    return skill_scores  # Return the skill scores array

# Scenario 2: Using pre-loaded nc_data variables directly (no file path)
def skill_score_scenario_no_error(nc_data_list, T_anom, MAAT_clumped_error, latties, longies):
    skill_scores = []  # List to store the skill scores
    for i, (latitudes_nc, longitudes_nc, mi_data) in enumerate(nc_data_list):
        mi = extract_mi_from_data(latitudes_nc, longitudes_nc, mi_data, latties, longies)
        skill_score = compute_skill_score_no_error(T_anom, MAAT_clumped_error, mi)
        skill_scores.append(skill_score)  # Store skill score in the list
       
    return skill_scores  # Return the skill scores array
```
